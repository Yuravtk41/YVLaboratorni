```python
# Програма, що вітає користувачів з іменами
імена = ["Олена", "Михайло", "Андрій", "Марія", "Ігор"]

for ім'я in імена:
    print(f"Привіт, {ім'я}! Приємно познайомитись!")
    вік = int(input(f"{ім'я}, скільки тобі років? "))

    # Умовні оператори для коментаря на основі віку
    if вік < 12:
        print(f"Вау, {ім'я}, ти ще досить молодий(а)! У тебе попереду багато цікавих пригод.")
    elif 12 <= вік < 18:
        print(f"{ім'я}, підліткові роки можуть бути дуже цікавими! Насолоджуйся цим часом.")
    elif 18 <= вік < 30:
        print(f"{ім'я}, найкращий час для нових можливостей і саморозвитку!")
    else:
        print(f"Вік – це лише цифра, {ім'я}! У будь-якому віці можна досягати нових вершин.")

    # Завершальне повідомлення
    print(f"Дякую за розмову, {ім'я}! Гарного дня!\n")
```

Ця програма вітає кількох користувачів із заздалегідь визначеними іменами. Для кожного користувача програма запитує його вік і надає персоналізовані відповіді на основі вікової категорії. Після завершення взаємодії з кожним користувачем програма дякує за розмову.

## Складові коду

- **Список імен**: Програма використовує список `імена`, який містить імена кількох користувачів: Олена, Михайло, Андрій, Марія, Ігор.
- **Цикл `for`**: Для кожного імені у списку програма виконує наступні дії:
    1. Виводить привітання з використанням імені.
    2. Запитує вік користувача.
    3. Використовує умовні оператори `if`, `elif`, `else`, щоб дати різні відповіді залежно від віку:

        - Якщо вік у межах від 12 до 18 років, програма дає пораду насолоджуватися підлітковими роками.
        - Для віку від 18 до 30 років підкреслюється важливість цього часу для саморозвитку.
        - Якщо вік користувача більше 30 років, програма каже, що вік — це лише цифра.
    4. Виводить завершальне повідомлення з подякою.
